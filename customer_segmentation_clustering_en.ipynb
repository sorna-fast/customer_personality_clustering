<li>
Remove outliers using IQR
</li>
<li>
<font color="red">Important Note:</font> It is not recommended to use the default values ​​for Q1 and Q3. This dataset has few data and removing a lot of data is not recommended. You can use lower/higher values ​​for Q1 and Q3. For example, 0.20 instead of 0.25 for Q1 and 0.80 for Q3. (Choose the values ​​yourself)
</li>
</font>
</div>
<br>

```python
# TODO: Remove outliers
def remove_outliers_iqr(df, columns, q1=0.20, q3=0.80):
    # This function removes outliers from specified columns of a DataFrame using the Interquartile Range (IQR) method.
    # df: The input pandas DataFrame.
    # columns: A list of column names from which to remove outliers.
    # q1: The quantile for the first quartile (default is 0.20 instead of the traditional 0.25).
    # q3: The quantile for the third quartile (default is 0.80 instead of the traditional 0.75).
    cleaned_df = df.copy() # Creates a copy of the DataFrame to avoid modifying the original.
    for col in columns:
        # Calculate quantiles with custom values
        Q1 = cleaned_df[col].quantile(q1) # Calculates the first quartile (Q1) for the current column.
        Q3 = cleaned_df[col].quantile(q3) # Calculates the third quartile (Q3) for the current column.
        IQR = Q3 - Q1 # Calculates the Interquartile Range (IQR).

        # Define the acceptable range
        lower_bound = Q1 - 1.5 * IQR # Calculates the lower bound for outlier detection.
        upper_bound = Q3 + 1.5 * IQR # Calculates the upper bound for outlier detection.

        # Filter out outliers
        mask = (cleaned_df[col] >= lower_bound) & (cleaned_df[col] <= upper_bound) # Creates a boolean mask to identify rows within the acceptable range.
        cleaned_df = cleaned_df[mask] # Filters the DataFrame, keeping only the rows that are not outliers.

    print(f"Number of deleted data: {len(df) - len(cleaned_df)}") # Prints the number of removed data points (outliers) in Persian.
    return cleaned_df # Returns the DataFrame with outliers removed.
```

```python
# Apply the function with q1=0.15 and q3=0.85
# Applies the 'remove_outliers_iqr' function to the DataFrame 'df'.
# It targets numerical columns (float64, int64, int32) and uses custom quantile values (q1=0.15, q3=0.85) for outlier detection.
df = remove_outliers_iqr(df, df.select_dtypes(include=['float64',"int64","int32"]).columns, q1=0.15, q3=0.85)
```

```python
# TODO: Complete the method below to display histograms and examine the mentioned variables.
def hist_plot(df, numeric_columns,
              bins=30, rot=None,
              figsize=(14, 8)):
    """
    This function displays histograms for numerical variables with the following settings:
    - Uses subplots to show multiple charts on one page.
    - Automatically adjusts rows/columns based on the number of variables.
    - Employs the seaborn color palette.
    - Allows for setting the rotation of X-axis labels.
    - Improves spacing between charts and labels.
    """

    # Initial settings
    n_cols = len(numeric_columns)
    if n_cols == 0:
        print("No numerical columns selected for plotting.") # Prints a message if no numerical columns are selected.
        return

    n_rows = (n_cols + 2) // 3  # Maximum 3 columns per row
    fig, axes = plt.subplots(n_rows, 3, figsize=figsize) # Creates a figure and a grid of subplots.
    axes = axes.flatten() # Flattens the 2D array of axes into a 1D array for easier iteration.
    sns.set_palette("husl")  # Sets the seaborn color palette.

    # Draw histograms for each numerical column
    for idx, col in enumerate(numeric_columns):
        if idx >= len(axes): # Checks if the number of numerical columns exceeds the available subplots.
            print(f"Warning: Number of numerical columns ({n_cols}) is greater than available axes ({len(axes)}). Some plots were not drawn.") # Issues a warning if not all plots can be drawn.
            break

        ax = axes[idx] # Gets the current subplot axis.

        # Ensure data is valid for plotting (e.g., no inf values)
        # This section can be further completed as needed for handling invalid data.
        # df_col_finite = df[col][np.isfinite(df[col])] # Only finite values (not inf and not NaN)
        # If you want NaN values to be removed as well:
        df_col_clean = df[col].dropna() # Removes NaN values from the column.
        df_col_clean = df_col_clean[np.isfinite(df_col_clean)] # Filters out infinite values.

        if df_col_clean.empty: # Checks if the cleaned column is empty.
            ax.text(0.5, 0.5, 'No valid data to plot', horizontalalignment='center', verticalalignment='center', transform=ax.transAxes) # Displays a message if no valid data is present.
            ax.set_title(f'Histogram of {col}', fontsize=10) # Sets the title for the empty plot.
            ax.set_xlabel(col, fontsize=8) # Sets the x-axis label.
            ax.set_ylabel('Frequency', fontsize=8) # Sets the y-axis label.
            if rot is not None:
                plt.setp(ax.get_xticklabels(), rotation=rot) # Rotates x-axis labels if 'rot' is specified.
            continue # Skips to the next column.

        # Draw histogram
        sns.histplot(
            data=df_col_clean, # Uses the cleaned data for plotting.
            # x=col, # If using data=df_col_clean, x=col is not needed as df_col_clean is a Series.
            bins=bins, # Sets the number of bins for the histogram.
            kde=True,  # Displays the Kernel Density Estimate (KDE) curve.
            ax=ax # Specifies the axes for the plot.
        )

        # Set title and labels
        ax.set_title(f'Histogram of {col}', fontsize=10) # Sets the title of the histogram.
        ax.set_xlabel(col, fontsize=8) # Sets the x-axis label of the histogram.
        ax.set_ylabel('Frequency', fontsize=8) # Sets the y-axis label of the histogram.

        # Set X-axis label rotation
        if rot is not None:
            plt.setp(ax.get_xticklabels(), rotation=rot, ha="right", rotation_mode="anchor") # Rotates x-axis labels for better alignment.

        # Display logarithmic scale for wide-ranging data (improved)
        # First, ensure there is data to calculate min/max
        if not df_col_clean.empty:
            min_val = df_col_clean.min() # Gets the minimum value in the cleaned column.
            max_val = df_col_clean.max() # Gets the maximum value in the cleaned column.
            # Only if min_val is positive and the ratio is reasonable
            if pd.notna(min_val) and pd.notna(max_val) and min_val > 0 and max_val > 0:
                if (max_val / min_val) > 1000: # Checks if the data range is very wide.
                    ax.set_yscale('log') # Sets the y-axis to a logarithmic scale.
            elif max_val > 100000 and min_val >=0 : # An alternative condition if min_val is zero but max_val is very large.
                     ax.set_yscale('log')


    # Hide extra axes that were not used
    for j in range(idx + 1, len(axes)): # Iterates through remaining unused subplots.
        axes[j].axis('off') # Turns off the axis for unused subplots.

    # Adjust spacing for better readability
    # h_pad: vertical space between plots
    # w_pad: horizontal space between plots
    # pad: overall padding from the figure edges
    plt.tight_layout(pad=1.5, h_pad=1.5, w_pad=1.0) # Adjusts subplot parameters for a tight layout. # You can change the values.

    # If spacing is still not optimal, you can use subplots_adjust:
    # fig.subplots_adjust(hspace=0.4, wspace=0.3) # Adjust values as needed

    plt.show() # Displays the plots.
```

```python
def Finding_the_right_number_of_clusters(model_cluster,pca_df):
    # Defines a range of cluster numbers to evaluate.
    cluster_range = range(2, 11)
    # Initializes lists to store silhouette and Davies-Bouldin scores.
    silhouette_scores = []
    davies_bouldin_scores = []

    # Iterates through each number of clusters in the defined range.
    for n_clusters in cluster_range:
        # Initializes the clustering model with the current number of clusters.
        agg_clustering = model_cluster(n_clusters=n_clusters)
        # Fits the model to the PCA-transformed data and predicts cluster labels.
        cluster_labels = agg_clustering.fit_predict(pca_df)
        # Calculates the silhouette score for the current clustering.
        sil_score = silhouette_score(pca_df, cluster_labels)
        silhouette_scores.append(sil_score) # Appends the silhouette score to its list.
        # Calculates the Davies-Bouldin score for the current clustering.
        db_score = davies_bouldin_score(pca_df, cluster_labels)
        davies_bouldin_scores.append(db_score) # Appends the Davies-Bouldin score to its list.

    # Creates a figure to display both plots.
    plt.figure(figsize=(12, 5))

    # Creates the first subplot for the Silhouette Score.
    plt.subplot(1, 2, 1)
    # Plots the silhouette scores against the number of clusters.
    plt.plot(cluster_range, silhouette_scores, marker='o')
    plt.title('Silhouette Score') # Sets the title for the Silhouette Score plot.
    plt.xlabel('Number of clusters') # Sets the x-axis label.
    plt.ylabel('Score') # Sets the y-axis label.

    # Creates the second subplot for the Davies-Bouldin Index.
    plt.subplot(1, 2, 2)
    # Plots the Davies-Bouldin scores against the number of clusters.
    plt.plot(cluster_range, davies_bouldin_scores, marker='o')
    plt.title('Davies-Bouldin Index') # Sets the title for the Davies-Bouldin Index plot.
    plt.xlabel('Number of clusters') # Sets the x-axis label.
    plt.ylabel('Score') # Sets the y-axis label.

    plt.tight_layout() # Adjusts subplot parameters for a tight layout.
    plt.show() # Displays the generated plots.
```

```python
# TODO: Find the right number of clusters for KMeans
# Calls the 'Finding_the_right_number_of_clusters' function.
# It passes the KMeans model as the clustering algorithm and 'pca_df' (the DataFrame with reduced dimensions) as the data.
# This function will compute and plot the Silhouette Score and Davies-Bouldin Index for a range of cluster numbers (2 to 10),
# helping to identify the optimal number of clusters for KMeans based on these internal validation metrics.
Finding_the_right_number_of_clusters(model_cluster=KMeans,pca_df=pca_df)
```

```python
def show_scatter_3d(model_cluster,pca_df,y):
    # Adds cluster labels to the PCA DataFrame, converting them to string type for categorical coloring.
    pca_df['Cluster'] = y.astype(str)

    # Tries to get cluster centers directly from the model (e.g., KMeans).
    try:
    # 2. Create DataFrame for cluster centers
        centers_df = pd.DataFrame(
            model_cluster.cluster_centers_, # Accesses the cluster centers attribute of the model.
            columns=['PC1', 'PC2', 'PC3'] # Assigns column names to the center coordinates.
        )
    except AttributeError:
        # If the model does not have a 'cluster_centers_' attribute (e.g., AgglomerativeClustering),
        # calculate centers manually as the mean of points in each cluster.
        centers = [] # Initializes an empty list to store calculated centers.
        for cluster in np.unique(y): # Iterates through each unique cluster label.
            cluster_points = pca_df[y == cluster][['PC1', 'PC2', 'PC3']] # Selects points belonging to the current cluster.
            cluster_center = cluster_points.mean(axis=0).values # Calculates the mean of the points to find the cluster center.
            centers.append(cluster_center) # Appends the calculated center to the list.
            centers_df = pd.DataFrame(centers, columns=['PC1', 'PC2', 'PC3']) # Creates a DataFrame from the calculated centers.

    centers_df['Cluster'] = 'Centroid' # Assigns a 'Centroid' label to the cluster centers for distinct visualization.

    # 3. Combine data with centers
    combined_df = pd.concat([pca_df, centers_df], ignore_index=True) # Concatenates the PCA DataFrame and the cluster centers DataFrame.

    # 4. Draw interactive 3D plot
    fig = px.scatter_3d(
        combined_df, # Data source for the plot.
        x='PC1',     # Column for the x-axis.
        y='PC2',     # Column for the y-axis.
        z='PC3',     # Column for the z-axis.
        color='Cluster', # Colors points based on the 'Cluster' column.
        symbol='Cluster', # Uses different symbols for points based on the 'Cluster' column (distinguishes centroids).
        size=[30]*len(pca_df) + [200]*len(centers_df), # Sets different sizes for data points and centroids.
        opacity=0.8, # Sets the opacity of the markers.
        title=f'3D {model_cluster} Clustering of Drivers', # Sets the title of the plot.
        labels={'PC1': 'Principal Component 1',
                'PC2': 'Principal Component 2',
                'PC3': 'Principal Component 3'}, # Custom labels for the axes.
        category_orders={'Cluster': sorted(pca_df['Cluster'].unique()) + ['Centroid']}, # Defines the order of categories in the legend.
        hover_data={'Cluster': True} # Specifies data to show on hover.
    )

    # 5. Visual settings
    fig.update_traces(
        marker=dict(line=dict(width=1, color='DarkSlateGrey')), # Sets marker border properties.
        selector=dict(mode='markers') # Applies settings to marker traces.
    )

    # 6. Layout settings
    fig.update_layout(
        legend_title_text='Clusters', # Sets the title for the legend.
        scene=dict(
            xaxis_title='PC1', # Sets x-axis title in 3D scene.
            yaxis_title='PC2', # Sets y-axis title in 3D scene.
            zaxis_title='PC3'  # Sets z-axis title in 3D scene.
        ),
        width=1000, # Sets the width of the plot.
        height=800 # Sets the height of the plot.
    )

    # 7. Display plot
    fig.show() # Displays the interactive 3D plot.
```

```python
# Range of k to test
k_range = range(1, 11) # Defines the range of cluster numbers (k) to evaluate, from 1 to 10.
bic_scores = [] # Initializes an empty list to store Bayesian Information Criterion (BIC) scores.
aic_scores = [] # Initializes an empty list to store Akaike Information Criterion (AIC) scores.
silhouette_scores = [] # Initializes an empty list to store Silhouette scores.

for k in k_range: # Iterates through each number of components (k) in the defined range.
    # Train GMM model
    gmm = GaussianMixture(n_components=k, covariance_type='full', random_state=42) # Initializes a GaussianMixture Model (GMM) with the current 'k' components, 'full' covariance type, and a fixed random state.
    gmm.fit(pca_result) # Fits the GMM to the PCA-transformed data.

    # Calculate metrics
    bic_scores.append(gmm.bic(pca_result)) # Appends the BIC score for the current GMM model to the list.
    aic_scores.append(gmm.aic(pca_result)) # Appends the AIC score for the current GMM model to the list.

    # Calculate Silhouette only for k > 1
    if k > 1: # Silhouette score is not defined for a single cluster.
        labels = gmm.predict(pca_result) # Predicts cluster labels using the fitted GMM.
        silhouette_scores.append(silhouette_score(pca_result, labels)) # Appends the Silhouette score to the list.
    else:
        silhouette_scores.append(np.nan) # Appends NaN for k=1 as Silhouette score is not applicable.

# Plot metric charts
plt.figure(figsize=(15, 5)) # Creates a figure to hold the subplots.

# BIC plot
plt.subplot(131) # Creates the first subplot in a 1x3 grid.
plt.plot(k_range, bic_scores, 'bo-') # Plots BIC scores against 'k', using blue circles and a solid line.
plt.xlabel('Number of Components (k)') # Sets the x-axis label.
plt.ylabel('BIC Score') # Sets the y-axis label.
plt.title('BIC for Optimal k') # Sets the title of the subplot.
plt.grid(True) # Displays a grid on the subplot.

# AIC plot
plt.subplot(132) # Creates the second subplot.
plt.plot(k_range, aic_scores, 'go-') # Plots AIC scores against 'k', using green circles and a solid line.
plt.xlabel('Number of Components (k)') # Sets the x-axis label.
plt.ylabel('AIC Score') # Sets the y-axis label.
plt.title('AIC for Optimal k') # Sets the title of the subplot.
plt.grid(True) # Displays a grid on the subplot.

# Silhouette plot
plt.subplot(133) # Creates the third subplot.
plt.plot(k_range[1:], silhouette_scores[1:], 'ro-') # Plots Silhouette scores (excluding k=1) against 'k', using red circles and a solid line.
plt.xlabel('Number of Components (k)') # Sets the x-axis label.
plt.ylabel('Silhouette Score') # Sets the y-axis label.
plt.title('Silhouette for Optimal k') # Sets the title of the subplot.
plt.grid(True) # Displays a grid on the subplot.

plt.tight_layout() # Adjusts subplot parameters for a tight layout.
plt.show() # Displays the generated plots.

# Select optimal k (lowest BIC)
optimal_k = np.argmin(bic_scores) + 1 # Finds the index of the minimum BIC score and adds 1 to get the corresponding 'k' value (since range starts from 1).
print(f"Optimal number of clusters (based on BIC): {optimal_k}") # Prints the optimal number of clusters determined by BIC.
```
